f(x) =  5*x**2 + 10*x - 300

Построить график

```python
from symtable import Symbol
from matplotlib.pyplot import plot
from sympy import *
from sympy import diff
x = Symbol('x')
f =  5*x**2 + 10*x - 300
plot(f)
print(f)
```

Определить корни

```python
roots = solve(f, x)
roots
for i in roots:
    print(float(i))
```

Найти интервалы, на которых функция возрастает

```python
print(diff(f))
d= diff(f)
solve(d>0,x)
```

Найти интервалы, на которых функция убывает

```python
solve(d<0, x)
```

Вычислить вершину

```python
def p(a, b, c):
    x0 = -b / (2 * a)
    y0 = ((4 * a * c) - (b * b)) / (4 * a)
    print(f'координата вершины: ({x0},{y0})')

p(5, 10, -300)
```

Определить промежутки, на котором f > 0

```python
solve(f>0)
```

Определить промежутки, на котором f < 0

```python
solve(f<0)
```
